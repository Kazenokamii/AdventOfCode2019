```python
def intCode(codes,inOne,inTwo,k=0,initial=0):
    x = codes.copy()
    while k < len(x):
        #print(k)
        #wait = input("PRESS ENTER TO CONTINUE.")
        if x[k] == 99:
            print('DONE')
            return 0,k,x,inTwo
        operator = str(x[k])
        if len(operator) == 1:
            if operator == '1':
                inputOne = x[x[k+1]]
                inputTwo = x[x[k+2]]
                outputPosition = x[k+3]
                x[outputPosition] = inputOne + inputTwo
                k = k+4
            elif operator == '2':
                inputOne = x[x[k+1]]
                inputTwo = x[x[k+2]]
                outputPosition = x[k+3]
                x[outputPosition] = inputOne * inputTwo
                k = k+4
            elif operator == '3':
                #inputOne = int(input("Enter an Integer: "))
                if initial:
                    inputOne = inTwo
                else:
                    inputOne = inOne
                    initial = initial+1
                outputPosition = x[k+1]
                x[outputPosition] = inputOne
                k = k+2
            elif operator == '4':
                displayOutput = x[x[k+1]]
                #print('Test Value should be 0: ' + str(displayOutput))
                k = k+2
                return 1,k,x,displayOutput
            elif operator == '5':
                inputOne = x[x[k+1]]
                inputTwo = x[x[k+2]]
                if inputOne != 0:
                    k = inputTwo
                else:
                    k = k+3
            elif operator == '6':
                inputOne = x[x[k+1]]
                inputTwo = x[x[k+2]]
                if inputOne == 0:
                    k = inputTwo
                else:
                    k = k+3
            elif operator == '7':
                inputOne = x[x[k+1]]
                inputTwo = x[x[k+2]]
                if inputOne < inputTwo:
                    x[x[k+3]] = 1
                else:
                    x[x[k+3]] = 0
                k = k+4
            elif operator == '8':
                inputOne = x[x[k+1]]
                inputTwo = x[x[k+2]]
                if inputOne == inputTwo:
                    x[x[k+3]] = 1
                else:
                    x[x[k+3]] = 0
                k = k+4
        else:
            modeOperator = str(x[k])[-2:]
            if modeOperator == '01':
                if len(operator) == 3:
                    inputOneMode = str(x[k])[-3]
                    inputTwoMode = 0
                elif len(operator) == 4:
                    inputOneMode = str(x[k])[-3]
                    inputTwoMode = str(x[k])[-4]
                if inputOneMode == '1':
                    inputOne = x[k+1]
                else:
                    inputOne = x[x[k+1]]
                if inputTwoMode == '1':
                    inputTwo = x[k+2]
                else:
                    inputTwo = x[x[k+2]]
                x[x[k+3]] = inputOne + inputTwo
                k = k+4
            elif modeOperator == '02':
                if len(operator) == 3:
                    inputOneMode = str(x[k])[-3]
                    inputTwoMode = 0
                elif len(operator) == 4:
                    inputOneMode = str(x[k])[-3]
                    inputTwoMode = str(x[k])[-4]
                if inputOneMode == '1':
                    inputOne = x[k+1]
                else:
                    inputOne = x[x[k+1]]
                if inputTwoMode == '1':
                    inputTwo = x[k+2]
                else:
                    inputTwo = x[x[k+2]]
                x[x[k+3]] = inputOne * inputTwo
                k = k+4
            elif modeOperator == '03':
                #inputOne = int(input("Enter an Integer: "))
                if initial:
                    inputOne = inTwo
                else:
                    inputOne = inOne
                    initial = initial+1
                outputPosition = x[k+1]
                x[outputPosition] = inputOne
                k = k+2
            elif modeOperator == '04':
                inputOneMode = inputOneMode = str(x[k])[-3]
                k = k+2
                if inputOneMode == '1':
                    #print('Test Value should be 0: ' + str(x[k+1]))
                    return 1,k,x,x[k+1]
                else:
                    #print('Test Value should be 0: ' + str(x[x[k+1]]))
                    return 1,k,x,x[x[k+1]]
            elif modeOperator == '05':
                if len(operator) == 3:
                    inputOneMode = str(x[k])[-3]
                    inputTwoMode = 0
                elif len(operator) == 4:
                    inputOneMode = str(x[k])[-3]
                    inputTwoMode = str(x[k])[-4]
                if inputOneMode == '1':
                    inputOne = x[k+1]
                else:
                    inputOne = x[x[k+1]]
                if inputTwoMode == '1':
                    inputTwo = x[k+2]
                else:
                    inputTwo = x[x[k+2]]
                if inputOne != 0:
                    k = inputTwo
                else:
                    k = k+3
            elif modeOperator == '06':
                if len(operator) == 3:
                    inputOneMode = str(x[k])[-3]
                    inputTwoMode = 0
                elif len(operator) == 4:
                    inputOneMode = str(x[k])[-3]
                    inputTwoMode = str(x[k])[-4]
                if inputOneMode == '1':
                    inputOne = x[k+1]
                else:
                    inputOne = x[x[k+1]]
                if inputTwoMode == '1':
                    inputTwo = x[k+2]
                else:
                    inputTwo = x[x[k+2]]
                if inputOne == 0:
                    k = inputTwo
                else:
                    k = k+3
            elif modeOperator == '07':
                if len(operator) == 3:
                    inputOneMode = str(x[k])[-3]
                    inputTwoMode = 0
                elif len(operator) == 4:
                    inputOneMode = str(x[k])[-3]
                    inputTwoMode = str(x[k])[-4]
                if inputOneMode == '1':
                    inputOne = x[k+1]
                else:
                    inputOne = x[x[k+1]]
                if inputTwoMode == '1':
                    inputTwo = x[k+2]
                else:
                    inputTwo = x[x[k+2]]
                if inputOne < inputTwo:
                    x[x[k+3]] = 1
                else:
                    x[x[k+3]] = 0
                k = k+4
            elif modeOperator == '08':
                if len(operator) == 3:
                    inputOneMode = str(x[k])[-3]
                    inputTwoMode = 0
                elif len(operator) == 4:
                    inputOneMode = str(x[k])[-3]
                    inputTwoMode = str(x[k])[-4]
                if inputOneMode == '1':
                    inputOne = x[k+1]
                else:
                    inputOne = x[x[k+1]]
                if inputTwoMode == '1':
                    inputTwo = x[k+2]
                else:
                    inputTwo = x[x[k+2]]
                if inputOne == inputTwo:
                    x[x[k+3]] = 1
                else:
                    x[x[k+3]] = 0
                k = k+4
    return x
```
